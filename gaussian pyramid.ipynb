```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img=cv2.imread("image.jpg",0)
data=np.asarray(img)
plt.imshow(data)
```

```python
data.shape
```

```python
ans=[]
for i in range(0,340):
    temp=[]
    for j in range(300):
        if j%4==0 and i%4==0:
            temp.append(data[i][j])
    if len(temp):
        ans.append(temp)
```

```python
plt.imshow(ans)
```

```python
ans1=[]
for i in range(len(ans)):
    temp=[]
    for j in range(len(ans[i])):
        if j%4==0 and i%4==0:
            temp.append(ans[i][j])
    if len(temp):
        ans1.append(temp)
```

```python
plt.imshow(ans1)
```

```python
ans2=[]
for i in range(len(ans1)):
    temp=[]
    for j in range(len(ans1[i])):
        if j%4==0 and i%4==0:
            temp.append(ans1[i][j])
    if len(temp):
        ans2.append(temp)
```
