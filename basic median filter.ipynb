```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=cv2.imread("image.jpg",0)
```

```python
plt.imshow(data)
```

```python
data.shape
```

```python
size=5
ans=[]
for i in range(225-size+1):
    temp=[]
    for j in range(225-size+1):
        temp1=[]
        for k in range(size):
            for l in range(size):
                temp1.append(data[i+k][j+l])
        temp1=sorted(temp1)
        median=temp1[len(temp1)//2]
        temp.append(median)
    ans.append(temp)
```

```python
ans=np.asarray(ans)
plt.imshow(ans)
```
