```python
import os
import cv2
path="usimages/"
files=os.listdir(path)
x=[]
y=[]
for i in files:
    if i.endswith(".png"):
        temp=cv2.imread(path+i,0)
        x.append(temp)
        y.append(i[:-4])

print(len(x))
print(len(y))
```

```python
import pytesseract
from PIL import Image
for i in range(len(x)):
    text = pytesseract.image_to_string(x[i])
    print(text,y[i])
```
