```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=cv2.imread("mona.jpg",0)
plt.imshow(data)
```

```python
size=5
[height,width] = data.shape
ans=[]
for i in range(height-size+1):
    temp=[]
    for j in range(width-size+1):
        temp1=[]
        for k in range(size):
            for l in range(size):
                temp1.append(data[i+k][j+l])
        temp1=sorted(temp1)
        median=temp1[len(temp1)//2]
        temp.append(median)
    ans.append(temp)
```

```python
ans=np.asarray(ans)
cv2.imwrite('monafil2.jpg',ans)
plt.imshow(ans)
```
