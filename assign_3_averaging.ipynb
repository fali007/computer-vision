```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("image.jpg",0)
data=np.asarray(img)
plt.imshow(data)
```

```python
data.shape
```

```python
i=11
fill=[[1/(i*i) for j in range(i)]for k in range(i)]
fill=np.asarray(fill)
```

```python
new=[]
for j in range(300-i+1):
    temp=[]
    for k in range(340-i+1):
        temp1=0
        for l in range(i):
            for m in range(i):
                temp1+=fill[l][m]*data[k+l][j+m]
        temp.append(temp1)
    new.append(temp)
new=np.transpose(new)
```
