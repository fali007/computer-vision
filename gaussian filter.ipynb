```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("image.jpg",0)
data=np.asarray(img)
plt.imshow(data)
```

```python
data.shape
```

```python
def gausian(size,sig):
    filtr=[[1/(2*3.14*sig**2)*np.exp(-(x**2+y**2)/(2*sig**2)) for x in range(size)]for y in range(size)]
    return filtr
```

```python
fill = [[1 / 256, 4  / 256,  6 / 256,  4 / 256, 1 / 256],
        [4 / 256, 16 / 256, 24 / 256, 16 / 256, 4 / 256],
        [6 / 256, 24 / 256, 36 / 256, 24 / 256, 6 / 256],
        [4 / 256, 16 / 256, 24 / 256, 16 / 256, 4 / 256],
        [1 / 256, 4  / 256,  6 / 256,  4 / 256, 1 / 256]]
```

```python
new=[]
for j in range(300-len(fill)+1):
    temp=[]
    for k in range(340-len(fill)+1):
        temp1=0
        for l in range(len(fill)):
            for m in range(len(fill)):
                temp1+=fill[l][m]*data[k+l][j+m]
        temp.append(temp1)
    new.append(temp)
new=np.transpose(new)
```

```python
plt.imshow(new)
```
