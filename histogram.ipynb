```python
import cv2
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("1537462354634-33f99e22-573c-4ccc-b8ca-a01e84e2ce10_.jpg")
```

```python
plt.imshow(img)
```

```python
img.shape
```

```python
a=[0 for i in range(256)]
b=[0 for i in range(256)]
c=[0 for i in range(256)]
for i in range(2000):
    for j in range(1500):
        a[img[i][j][0]]+=1
        b[img[i][j][1]]+=1
        c[img[i][j][2]]+=1
print(a,b,c)
```

```python
t=[i for i in range(256)]
plt.plot(t,a,'b',t,b,'g',t,c,'r')
plt.plot(b)
plt.show()
```

```python
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```

```python
a_=[a[i]/(280*414) for i in range(256)]
b_=[b[i]/(280*414) for i in range(256)]
c_=[c[i]/(280*414) for i in range(256)]
```

```python
a_=np.asarray(a_)
b_=np.asarray(b_)
c_=np.asarray(c_)
print(a_.mean(),b_.mean(),c_.mean())
```

```python
mean=[0.10109342650103519 for i in range(256)]
plt.plot(t,a_,'b',t,b_,'g',t,c_,'r',mean)
plt.show()
plt.savefig("histogram.pdf")
```

```python
# plt.plot(a_.mean(),'r',b_.mean(),'b',c_.mean(),'g')
# plt.show()
print(a_.mean(),a_.var(),np.median(a_))
print(b_.mean(),b_.var(),np.median(b_))
print(c_.mean(),c_.var(),np.median(c_))
```

```python
# eq=[eq[i]*255 for i in range(255)]
sum_a,sum_b,sum_c=0,0,0
cdf_a,cdf_b,cdf_c=[],[],[]
for i in range(len(a_)):
    sum_a+=a_[i]
    sum_b+=b_[i]
    sum_c+=c_[i]
    cdf_a.append(sum_a)
    cdf_b.append(sum_b)
    cdf_c.append(sum_c)
cdf_a=[i/max(cdf_a) for i in cdf_a]
cdf_b=[i/max(cdf_b) for i in cdf_b]
cdf_c=[i/max(cdf_c) for i in cdf_c]
# print(cdf_a,cdf_b,cdf_c)
```
