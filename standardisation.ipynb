```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
img=cv2.imread("image.jpg",0)
data=np.asarray(img)
plt.imshow(data)
```

```python
data.shape
```

```python
img1=cv2.imread("image1.jpg",0)
img1 = cv2.resize(img1, (300,340), interpolation = cv2.INTER_AREA)
data1=np.asarray(img1)
plt.imshow(data1)
```

```python
data1.shape
```

```python
ans=[[0 for i in range(300)]for j in range(340)]
for i in range(256):
    sum=0
    l=0
    for j in range(340):
        for k in range(300):
            if data1[j][k]==i:
                sum+=data[j][k]
                l+=1
    val=np.ceil(sum/l)
    for j in range(340):
        for k in range(300):
            if data1[j][k]==i:
                ans[j][k]=val
```

```python
ans=np.asarray(ans)
plt.imshow(ans)
ans
```
